## Research Question

How do seasonal weather fluctuations, as measured by average temperature, affect university students’ self reported happiness levels, as measured by scores on anxiety or depressive feelings questionnaire? Do these self reported happiness levels have an effect on student academic performance?

## Background and Prior Work

Weather is thought to have significant impacts on human well being. Psychologically this phenomenon can be explained by Maslow's Hierarchy of needs where reliable good weather conditions can satisfy an individual's basic physiological and safety needs <a name="cite_ref-1a"></a>[<sup>1</sup>](#cite_note-1). “The evolutionary theory posits that humans have an inherent positive preference for non-threatening environmental factors in nature, which helps to promote survival and enhance well-being” <a name="cite_ref-1b"></a>[<sup>1</sup>](#cite_note-1). Light and temperature show to play a role in people's physiological and psychological state, sufficient sunlight and warm temperature are often correlated with positive emotions and mental state <a name="cite_ref-1c"></a>[<sup>1</sup>](#cite_note-1). The association between weather and well-being is due to a combination of factors, including light, temperature, air quality and several other weather conditions. The positive effect of sunlight could be due to more accessibility to the outdoors which bolsters life satisfaction <a name="cite_ref-1d"></a>[<sup>1</sup>](#cite_note-1). Sunlight has direct impacts on the production of melatonin, a hormone responsible for the regulation of the circadian rhythm and neurotransmitters associated with happiness and well-being like serotonin, dopamine, and endorphins <a name="cite_ref-1e"></a>[<sup>1</sup>](#cite_note-1). 

A study <a name="cite_ref-2a"></a>[<sup>2</sup>](#cite_note-2) on SAD in students examined whether a college student’s risk of developing symptoms of seasonal depressive disorder is higher when they moved from a home town with minimal seasonal weather fluctuations (like Hawaii) to a university located in an area of high seasonality (like Pennsylvania). The authors hypothesized that students raised in non-seasonal environments would show larger winter increases in depressive symptoms compared to students from highly seasonal hometowns. Using an adapted version of the Seasonal Patterns Assessment Questionnaire (SPAQ) administered to 3 different universities in the fall (as baseline) and again in the winter and comparing that data to publicly available climate data, they found that nearly 50% of students from non-seasonal hometowns reported an increased SPAQ score in the winter, compared to no change for those with non-seasonal hometowns. Students from non-seasonal hometowns appear to experience greater seasonal shifts in mood and there are other adaptive factors that affect seasonal mood changes <a name="cite_ref-2b"></a>[<sup>2</sup>](#cite_note-2).

Another study <a name="cite_ref-3a"></a>[<sup>3</sup>](#cite_note-3) examined whether students’ happiness (subjective well-being) predicts academic performance, and whether gender differences exist during the disruption of the COVID-19 pandemic. They used The Oxford Happiness Questionnaire (OHQ) as a self reported measure of happiness and grade point average as a measure of academic success. Data analysis revealed that higher education students in the UAE have high levels of happiness and academic success with no gender difference, indicating that students’ happiness levels were positively correlated to their academic success <a name="cite_ref-3b"></a>[<sup>3</sup>](#cite_note-3). 
	
While many prior studies rely on surveys and psychological assessments, recent research has increasingly used large-scale behavioral datasets to examine how weather influences mood in real-world settings. Another study <a name="cite_ref-4a"></a>[<sup>4</sup>](#cite_note-4) of prior work examines weather influences on emotional states at scale using large datasets, linking environmental conditions to expressed sentiment and mood beyond traditional survey methods. For example, one large analysis of 3.5 billion social media posts from Facebook and Twitter found that multiple weather variables like temperature, precipitation, humidity, and cloud cover, were strongly associated with the sentiment expressed in people’s posts. Adverse weather conditions correspond to more negative emotional expression in its data <a name="cite_ref-4b"></a>[<sup>4</sup>](#cite_note-4). Similarly, a global study of 1.2 billion social media posts across 157 countries observed that extremely high temperatures were associated with more negative moods. In other words, climate variables were found to be related to mood at the global level <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). Recent research using computational methods has improved the way that language and regional variations are taken into account when examining the relationship between weather and sentiment on social media. The research points to the complex relationship that weather has with mood and the importance of using context-sensitive analysis when examining data <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6). These studies show that data science techniques are effective in examining the relationship that weather has with mood and that similar techniques could be used in our project.

## Hypothesis


We predict that as the average monthly temperature increases, self-reported levels of anxiety and depression will go down in university students (signifying lower levels of anxiety and depression). We also predict that as the average monthly temperature decreases, self-reported anxiety or depressive feelings questionnaire scores will increase (signifying higher levels of anxiety and depression). Lastly, we believe that a decrease in self reported anxiety and depression questionnaire scores will also result in a decrease in academic performance as measured by semester or quarter GPA averages. 

We believe this hypothesis because of the current accepted theory that sunlight has an effect on the circadian rhythm which in turn has an effect on mood regulation and by extension happiness (1) as well as the individual studies relating climate patterns and happiness in university students (2) and student self reported happiness and grade performance.

## Data

The ideal dataset for answering this question would include the temperatures of the universities we look at, spanning at least a year. If we choose to analyze the population of one university (like UCSD) then we would need temperature data from that specific area. It would be most ideal to get data for a precise zip code. We would ideally need a validated self reported happiness questionnaire to measure student happiness that is administered over multiple quarters/ semesters for comparison through time as well as the reported GPA or self reported academic performance associated with the same participant for every quarter/semester. Climate data could be found through publicly available NOAA records while happiness/academic performance data could be found in open source student quarterly/semester surveys. This data will be collected and stored in a tabular format and compared on a time scale to the average weather data of that time period (i.e. survey data from winter quarter will be compared to weather data from January to March. 

Potential datasets:
1. <a href="https://the-globem.github.io/datasets/overview">Globem University of Washington</a>
   * GLOBEM datasets contain the first released multi-year mobile and wearable sensing datasets that include four years of data collection studies (2018-2021), conducted at an R-1 university. To access the data, we need to: become a credentialed user of Physionet (includes references), CITI Data or Specimens Only Research Training, sign a data use agreement. From this dataset we can extract data from the weekly Ecological Momentary Assessment (EMA) surveys which capture participants' recent sense of their mental health, specifically responses to questions for anxiety and depression questionnaire.
2. <a href="https://www.ncei.noaa.gov/access/past-weather/">NOAA daily average temperatures</a>
   * This website provides daily average temperatures from the closest weather station to the inputted location. This data is available free to download in csv, json, xtml, or txt format with no need to request permission. The variables that would be useful in this data set would be the date and the average temperature on that date for the matching time period during which the GLOBEM study (listed above) was operational.
