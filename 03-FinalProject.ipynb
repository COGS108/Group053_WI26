## Research Question

How do daily weather fluctuations, as measured by average temperature, affect university students’ self reported depression and anxiety levels, as measured by scores on anxiety or depressive feelings questionnaire? Do daily weather fluctuations have an effect on the duration of sleep university students get per night? Is there an association between nightly sleep duration and self reported feelings of depression and anxiety in students? 

## Background and Prior Work

Weather is thought to have significant impacts on human well being. Psychologically this phenomenon can be explained by Maslow's Hierarchy of needs where reliable good weather conditions can satisfy an individual's basic physiological and safety needs (1). “The evolutionary theory posits that humans have an inherent positive preference for non-threatening environmental factors in nature, which helps to promote survival and enhance well-being” (1). Light and temperature show to play a role in people's physiological and psychological state, sufficient sunlight and warm temperature are often correlated with positive emotions and mental state (1). The association between weather and well-being is due to a combination of factors, including light, temperature, air quality and several other weather conditions. The positive effect of sunlight could be due to more accessibility to the outdoors which bolsters life satisfaction (1). Sunlight has direct impacts on the production of melatonin, a hormone responsible for the regulation of the circadian rhythm and neurotransmitters associated with happiness and well-being like serotonin, dopamine, and endorphins. (1).

A study (2) on SAD (seasonal affective disorder) in students examined whether a college student’s risk of developing symptoms of seasonal affective disorder is higher when they moved from a home town with minimal seasonal weather fluctuations (like Hawaii) to a university located in an area of high seasonality (like Pennsylvania). The authors hypothesized that students raised in non-seasonal environments would show larger winter increases in depressive symptoms compared to students from highly seasonal hometowns. Using an adapted version of the Seasonal Patterns Assessment Questionnaire (SPAQ) administered to 3 different universities in the fall (as baseline) and again in the winter and comparing that data to publicly available climate data, they found that nearly 50% of students from non-seasonal hometowns reported an increased SPAQ score in the winter, compared to no change for those with non-seasonal hometowns. Students from non-seasonal hometowns appear to experience greater seasonal shifts in mood and there are other adaptive factors that affect seasonal mood changes.(2).

A study on the effects of weather on sleep patten has shown a “small but statistically significant effects of seasonal and weather effects on sleep patterns” (3) finding that “the strongest seasonal effects for wake time and sleep duration, especially during the spring season: wake times are earlier, and sleep duration decreases (compared to the reference season winter)” (3). “Sleep duration also modestly decreases when day lengths get longer (between the winter and summer solstice). Bedtimes and wake times tend to be slightly later as outdoor temperature increases” (3). Several demographics were controlled for including age, sex, organization, supervisory role, latitude, and longitude of home location as well as psychological traits associated with sleep using the Positive and Negative Affect Schedule (PANAS X) for affect balance, Big Five Inventory (BFI) for personality, the Pittsburgh Sleep Quality Index (PSQI) for sleep quality, and Morning-Eveningness Questionnaire (MEQ)107 for chronotype (3). This study shows that there is a correlation between seasonality and sleep duration with longer day lengths associated with less nightly sleep duration. For every hour increase in daylight, sleep duration decreases by 3.6 minutes. However, “the addition of weather variables (temperature, wind, and humidity/cloud cover) in the full model did not explain significantly more variance” (3).

The effect of sleep duration on general health outcomes was also looked at in a study that took self reported sleep data from the a sample of US adults and categorized them into 3 groups: short sleep (≤5 hours), recommended sleep (6–8 hours), and long sleep (≥9 hours) (7). Health outcomes were measured in a variety of ways including self-reported diagnosis of depression, number of poor mental health days, number of poor physical health days, and self-rated general health, measured on a 5-point Likert scale from excellent to poor (7). The study found that short sleep duration (≤5 hours) was associated with a 14.1 percentage point increase in depression incidence (95% CI: 13.8%–14.4%), while long sleep duration (≥9 hours) was linked to a 12.9 percentage point increase (7). Both short and long sleep groups also reported having more poor mental and physical health days as well as lower self-rated general health (7). This study shows that there is a correlation between sleep duration and general mood and well-being, with the best outcomes coming from people who sleep on average 6 to 8 hours per night.

Another study (3) examined whether students’ happiness (subjective well-being) predicts academic performance, and whether gender differences exist during the disruption of the COVID-19 pandemic. They used The Oxford Happiness Questionnaire (OHQ) as a self reported measure of happiness and grade point average as a measure of academic success. Data analysis revealed that higher education students in the UAE have high levels of happiness and academic success with no gender difference, indicating that students’ happiness levels were positively correlated to their academic success. (3).

While many prior studies rely on surveys and psychological assessments, recent research has increasingly used large-scale behavioral datasets to examine how weather influences mood in real-world settings. Another study (4) of prior work examines weather influences on emotional states at scale using large datasets, linking environmental conditions to expressed sentiment and mood beyond traditional survey methods. For example, one large analysis of 3.5 billion social media posts from Facebook and Twitter found that multiple weather variables like temperature, precipitation, humidity, and cloud cover, were strongly associated with the sentiment expressed in people’s posts. Adverse weather conditions correspond to more negative emotional expression in its data (4). Similarly, a global study of 1.2 billion social media posts across 157 countries observed that extremely high temperatures were associated with more negative moods. In other words, climate variables were found to be related to mood at the global level (5). Recent research using computational methods has improved the way that language and regional variations are taken into account when examining the relationship between weather and sentiment on social media. The research points to the complex relationship that weather has with mood and the importance of using context-sensitive analysis when examining data (6). These studies show that data science techniques are effective in examining the relationship that weather has with mood and that similar techniques could be used in our project.

## Hypothesis


We predict that as the average weekly temperature increases, self-reported levels of anxiety and depression will go down in university students (signifying lower levels of anxiety and depression). We also predict that as the average weekly temperature decreases, self-reported anxiety or depressive feelings questionnaire scores will increase (signifying higher levels of anxiety and depression). We predict that lower average weekly temperatures will also increase sleep duration and that there is an association between sleep duration and self-reported depression and anxiety ratings.

We believe this hypothesis because of the current accepted theory that sunlight affects the circadian rhythm, which in turn has an effect on mood regulation and, by extension, levels of self-reported depression and anxiety (1) as well as the individual studies relating climate patterns and SAD rates in university students (2). We believe that there is a negative association between temperature and sleep patterns, with lower temperatures increasing average sleep duration because lower temperatures limit the appeal and availability of outside activities and are usually associated with “poor weather,” which increases the likelihood of students staying inside and therefore provides more opportunity for sleep. Lastly, we believe that there will be an association with sleep times and self-reported depression and anxiety ratings due to studies (2) findings that people who sleep more or less than 6-8 hours report higher levels of depression.

### Dataset #2: Weather

```python
# 2019 Weather Data
w19 = pd.read_csv('data/00-raw/weather-2019.csv') 

# Change column names and clear first two rows.
w19_tidy = w19.rename(columns={'latitude': 'date', 'longitude': 'avg_temperature_F'})
w19_tidy = w19_tidy.iloc[2:]

w19_tidy.to_csv('data/01-interim/weather-2019_tidied.csv',index=False)

# Check the size of the dataset.
print(f"\nDataset Shape: {w19_tidy.shape}")

# Count missing data.
missing_counts = w19_tidy.isnull().sum()

# Check for outliers and suspicious entries.
w19_tidy['avg_temperature_F'] = pd.to_numeric(w19_tidy['avg_temperature_F'], errors='coerce')
q1 = w19_tidy['avg_temperature_F'].quantile(0.25)
q3 = w19_tidy['avg_temperature_F'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = w19_tidy[(w19_tidy['avg_temperature_F'] < lower_bound) | (w19_tidy['avg_temperature_F'] > upper_bound)]
w19_tidy['is_outlier'] = (w19_tidy['avg_temperature_F'] < lower_bound) | (w19_tidy['avg_temperature_F'] > upper_bound)

print(f"\nNumber of statistical outliers found: {len(outliers)}")

# Clean the data.
#w19_clean = w19_tidy[w19_tidy['is_outlier'] == False]
w19_clean = w19_tidy[['date','avg_temperature_F']].copy()

# Change hourly data to daily data.
w19_clean['date'] = pd.to_datetime(w19_clean['date'])
w19_clean = w19_clean.set_index('date')
w19_clean = w19_clean.resample('D').mean().reset_index()
w19_clean['avg_temperature_F'] = w19_clean['avg_temperature_F'].round(1)
w19_clean = w19_clean.infer_objects(copy=False)

print("\nCleaned dataset:\n",w19_clean.head())
w19_clean.to_csv('data/02-processed/weather-2019_cleaned.csv',index=False)
```

### Exploratory Data Analysis

#### Overview
This exploratory analysis was meant to examine the relationship between weekly average temperature, sleep duration, and self-reported anxiety/ depression levels (using PHQ-4 scores) in University of Washington undergraduate students. The Following tests evaluate:
1) Whether weekly temperature fluctuations are associated with PHW-4 scores
2) Whether temperature influences sleep duration
3) Whether sleep duration is associated with PHQ-4 scores.
4) Whether sleep mediates the relationship between temperature and mental health.

#### Descriptive Patterns

##### Temperature and PHQ-4
Initial correlations and linear regression suggested a small positive association between temperature and PHQ-4 scores. A warmer average weekly temperature is associated with slightly higher anxiety/ depression scores. However, the effect size was small. 

##### Temperature and Sleep Duration

Mixed-effects modeling showed that higher temperatures were significantly associated with reduced sleep duration
- For each 1°F increase, sleep decreased by approximately 0.117 hours per week (~7 minutes per week)
- A 10°F increase corresponded to approximately 1.17 fewer hours of sleep per week.

This finding supports the original hypothesis that colder weekly temperatures are associated with longer sleep duration. 

##### Sleep and PHQ-4 Scores
Sleep duration showed a small negative association with PHQ-4 scores. More sleep corresponded to slightly lower anxiety and depression scores, but this relationship did not withstand more specified models. 

##### Between-person variability
There were substantial between-person differences, which we decomposed into within-person effects (weekly deviations from participants' average) and between-person effects (participant-level averages). This allowed us to test whether short term fluctions in temperature affected PHQ-4 scores, independent of individual differences. The findings from these tests were as follows: 
- Within-person temperature deviations were significant predictors of PHQ-4 scores.
    -When the temperature was higher than the student's usual weekly average, PHQ-4 scores were slightly elevated
- Between-person temperature differences were not significant
- Neither within nor between-person sleep duration significantly predicted PHQ-2 scores in the fully specified model.

##### Effect sizes
Although temperature effects were statistically significant, the effect sizes were small.
- A 10°F weekly increase corresponds to approximately a 0.26-point increase in PHQ-4.
A large sample size increases statistical power, meaning small effects can be statistically significant however, limited in practical significance.

##### Summary of Findings
1. Weekly temperature fluctuations are associated with small changes in PHQ-4 scores.
3. Higher temperatures are associated with reduced sleep duration.
4. Sleep does not strongly mediate the relationship between temperature and PHQ-4.
5. Between-person variability in mental health is substantially larger than temperature-driven fluctuations.

The EDA shows that short-term temperature fluctuation may influence student mental health but individual differences account for the majority of variance.

```python
def correlation_summary(df, x, y):
    sub = df[[x, y]].dropna()

    pearson_r, pearson_p = stats.pearsonr(sub[x], sub[y])
    spearman_rho, spearman_p = stats.spearmanr(sub[x], sub[y])

    return pd.Series({
        "N": len(sub),
        "Pearson r": pearson_r,
        "Pearson p": pearson_p,
        "Spearman rho": spearman_rho,
        "Spearman p": spearman_p
    })
```

#### Random-effect ANOVA
To quantify participant-level clustering, we first fit a null random-intercept model with participant ID as the grouping factor. This model serves as a random-effects ANOVA analogue and allows us to estimate the intraclass correlation coefficient (ICC). The ICC represents the proportion of variance in PHQ-4 attributable to stable between-person differences rather than week-to-week fluctuations within a person.

The estimated ICC from this model was 0.71, indicating that approximately 71% of the variance in PHQ-4 scores occurs between participants rather than within participants across weeks. This large ICC suggests that individuals differ substantially in their baseline PHQ-4 levels, providing strong justification for the use of a mixed-effects model that accounts for participant-level clustering.

```python
# Null random-intercept model
null_mixed = smf.mixedlm(
    "phq4 ~ 1",
    data=analysis_df,
    groups=analysis_df["pid"]
).fit(reml=True)

print(null_mixed.summary())

# Intraclass Correlation Coefficient (ICC)
random_var = null_mixed.cov_re.iloc[0, 0]
resid_var = null_mixed.scale
icc = random_var / (random_var + resid_var)

print("Random intercept variance:", random_var)
print("Residual variance:", resid_var)
print("ICC:", icc)
```

#### Why a Mixed-Effects Model?

The dataset contains repeated weekly observations from the same participants. Because multiple rows come from the same individual, the observations are not independent. Standard linear regression assumes that all observations are independent. Because weekly observations from the same participant are likely correlated, this assumption is violated, which can lead to underestimated standard errors.

To account for this structure, we use a mixed-effects model with a random intercept for participant ID. This allows each participant to have their own baseline PHQ-4 level while estimating the average fixed effects of temperature and sleep across the sample. Mixed-effects models are therefore appropriate for longitudinal or repeated-measures datasets such as this one.

#### Random Intercept Mixed Model


$$
\text{PHQ-4}_{ij} = \beta_0 + \beta_1(\text{Temperature}_{ij}) + \beta_2(\text{Sleep}_{ij}) + u_{0j} + \varepsilon_{ij}
$$

Where:

- $i$ = week  
- $j$ = participant  
- $u_{0j}$ = random intercept for participant $j$  
- $\varepsilon_{ij}$ = residual error

We next fit a random-intercept mixed-effects model predicting PHQ-4 from weekly temperature and weekly sleep duration while accounting for repeated observations nested within participants. Year was included as a fixed effect to adjust for potential differences across 2019–2021.

In this model, average weekly temperature remained a statistically significant but very small positive predictor of PHQ-4 scores (β = 0.026, p < .001). Sleep duration showed a small negative coefficient but was not statistically significant (β = −0.007, p = .067). The fixed effects for year were also not statistically significant, suggesting that average PHQ-4 levels did not differ meaningfully across the three years in this model.

The random intercept variance for participant ID remained large, indicating that substantial variation in PHQ-4 scores reflects stable differences between participants rather than week-to-week changes within individuals. This finding is consistent with the high intraclass correlation observed in the null model and supports the use of a mixed-effects modeling framework for this dataset.



```python
mixed_phq = smf.mixedlm(
    "phq4 ~ avg_temperature_F + sleep_hours + C(year)",
    data=analysis_df,
    groups=analysis_df["pid"]
).fit(reml=False)

print(mixed_phq.summary())
```

\text{Sleep}_{ij} = \beta_0 + \beta_1(\text{Temperature}_{ij}) + u_{0j} + \varepsilon_{ij}
$$

Where:

- $\beta_0$ = overall average sleep duration

#### Within and Between-Person Effects of Temperature and Sleep on PHQ-4 Scores
To distinguish week-to-week changes within the same individual from stable differences between individuals, we fit a mixed-effects model using person-mean centering. The within-person temperature term tests whether a participant reports different PHQ-4 scores during warmer-than-usual weeks relative to their own average temperature, while the between-person term tests whether participants who are generally exposed to warmer temperatures differ in PHQ-4 scores from other participants overall.

The results show that within-person temperature was a statistically significant predictor of PHQ-4 scores (β = 0.026, p < .001), indicating that during weeks when temperatures were higher than a participant’s typical level, their PHQ-4 scores tended to be slightly higher. In contrast, the between-person temperature effect was not statistically significant (β = 0.072, p = .220), suggesting that individuals who generally experienced warmer conditions did not differ meaningfully in PHQ-4 scores from those who experienced cooler conditions overall.

Both the within-person and between-person sleep effects were not statistically significant (p = .124 and p = .156, respectively), indicating that weekly sleep duration did not meaningfully predict PHQ-4 scores after separating within- and between-person variation.

The random intercept variance remained large (Group Var = 5.85), indicating substantial differences in baseline PHQ-4 scores across participants. This pattern is also visible in the distribution of average PHQ-4 scores across students, where scores vary widely between individuals. Together, these results suggest that most variation in PHQ-4 scores is driven by stable differences between participants, while temperature shows only a small week-to-week association within individuals.

## Discusison and Conclusion

This study examines whether weekly temperature fluctuations are associated with sleep duration and self-reported anxiety and depression. Studies also show that longer daylight hours are associated with slightly shorter sleep duration, and both very short and very long sleep are linked to poorer mental and physical health. Because of the idea that sunshine influences the circadian rhythm, which in turn impacts mood regulation and, consequently, levels of self-reported anxiety and depression, we believe there is a relationship between temperature and sleep patterns. In addition, extensive investigations of social media data have revealed a relationship between weather and people's reported mood, with extreme or unfavorable weather frequently being connected with more negative sentiment. Therefore, we also want to examine the relationship between sleep duration and self-reported depression and anxiety ratings. We hypothesize that there will be a negative correlation between temperature and sleep patterns, as well as a correlation between sleep duration and self-reported anxiety and depression scores.

We began by cleaning the dataset and performing some descriptive analyses. For our analysis, we performed simple linear regression and ordinary least squares (OLS) regression to look for the bivariate associations between Weekly Temperature and PHQ-4, Weekly Sleep Duration and Weekly Temperature, and Weekly Sleep Duration and PHQ-4 Scores. The results show a small negative correlation between weekly sleep duration and PHQ-4 scores, a small positive correlation between average weekly temperature and PHQ-4 scores, and a small negative association between average weekly temperature and weekly sleep duration. Then, we performed multiple linear regression using OLS to examine whether weekly temperature and weekly sleep duration predicted PHQ-4 scores.The overall model was statistically significant, though the variance explained was small. Higher temperatures were associated with slightly higher PHQ-4 scores. Greater sleep duration was associated with slightly lower PHQ-4 scores. However, the overall effect sizes were very small, indicating that temperature and sleep duration explain only a small proportion of variation in PHQ-4 scores. This suggests that other factors beyond weather conditions may play a larger role in influencing students’ mental health.

In the last part of our project, we estimated a mixed-effects model with a random intercept for temperature and sleep to account for repeated observations within participants. The model revealed that average weekly temperature was statistically significantly associated with weekly sleep duration, although the effect size was small. Higher temperatures were associated with slightly reduced sleep duration, with each 1°F increase corresponding to approximately 0.12 fewer hours of sleep per week. Substantial between-person variability in baseline sleep duration remained, suggesting that stable individual differences account for much more variance in sleep than weekly temperature fluctuations.

Overall, we partially support our hypothesis that there is a significant relationship between weekly temperature fluctuations and PHQ-4 scores, as well as a negative correlation between temperatures and sleep duration. However, we fail to reject the null hypothesis that there is no association between sleep duration and self-reported depression and anxiety ratings, as sleep does not strongly mediate the relationship between temperature and PHQ-4. According to the analysis, individual differences account for most of the variance, with between-person variability being significantly greater than temperature-driven changes, even if short-term temperature fluctuations may have an impact on students' mental health.

There are also some limitations that need to be considered. When we tried to analyze the GLOBEM dataset, we found that there are many missing values in the 2019 dataset, which may affect the accuracy of our estimations. Additionally, weather data collected from certain stations may not exactly represent the environmental conditions experienced by individuals in the behavioral dataset. In addition, the dataset consists only of university students, which may limit the generalizability of these findings to other populations. Furthermore, although temperature effects were statistically significant, the effect sizes were small in magnitude and may lack practical significance.

Future studies might build on this topic by using larger datasets, using statistical or machine learning models, and looking at other elements like seasonal trends or demographic variations in order to advance this research question and our project. These approaches would help clarify the strength and direction of the relationships observed and provide a more comprehensive understanding of how weather conditions may influence sleep, mood, and overall well-being. Overall, these findings suggest that while environmental conditions such as temperature may have measurable effects on sleep and mood, their influence appears to be modest compared to stable individual differences and other contextual factors.
